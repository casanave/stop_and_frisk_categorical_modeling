```python
import pandas as pd
```

```python
# a small function to remove the seconds
def format_twenty_and_nineteen(date):
    date = date[:-3]
    return date
```

```python
# a small function to format 2018's datetimes
def format_eighteen(date):         
    date = date[0:11] + date[20:-3]
    return date
```

```python
# a small function to format 2017's datetimes
def format_seventeen(date):
    date = date[:16]
    if 'nan' in date:
        date = date[:10] + ' 00:00:00'
    return date
```

```python
df_17 = pd.read_csv('raw_data/SF_2017.csv.gz')
df_17.rename({'STOP_FRISK_DATE_STOP_FRISK_TIME' : 'DATE_TIME'}, axis = 1, inplace = True)
df_17['DATE_TIME'] = df_17['DATE_TIME'].map(format_seventeen)
df_17.head()
```

```python
df_18 = pd.read_csv('raw_data/SF_2018.csv.gz')
df_18.rename({'STOP_FRISK_DATE_Stop Frisk Time' : 'DATE_TIME'}, axis = 1, inplace = True)
df_18['DATE_TIME'] = df_18['DATE_TIME'].map(format_eighteen)
df_18.head()
```
